```python
import numpy as np
```

```python
class LinearRegression:
    def __init__(self, leanringRate = 0.0001, num_iteration=10000):
        self.lr = leanringRate
        self.n_iters = num_iteration
        self.weights = None
        self.bias = None
    
    def fit(self, trainningSample, labelSample):
        #This function involves on training steps and graddient descent
        
        #init parameters
        number_samples,number_features = trainningSample.shape
        self.weights = np.zeros(number_features)
        self.bias = 0
        
        #gradient descent
        for _ in range(self.n_iters):
            y_predicted = np.dot(trainningSample, self.weights)+self.bias
            
            # Calculating the derivative
            dw = (1/number_samples) * np.dot(trainningSample.T,(y_predicted - labelSample))
            db = (1/number_samples) * np.sum(y_predicted-labelSample)
            
            #Updating our weights and bias
            self.weights -= self.lr * dw
            self.bias -= self.lr * db
            
    
    def predict(self, testSamples):
        #This function takes the new data samples and then it can approximate the value
        y_predicted = np.dot(testSamples, self.weights)+self.bias
        return y_predicted
    
```

```python
 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
import matplotlib.pyplot as plt


X, y = datasets.make_regression(n_samples=100, n_features=1, noise=20, random_state=4)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)

fig = plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], y, color = "b", marker = "o", s = 30)
plt.show()

```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
predictions = reg.predict(X_test)

def mse(y_test, predictions):
    return np.mean((y_test-predictions)**2)

mse = mse(y_test, predictions)
print(mse)

y_pred_line = reg.predict(X)
cmap = plt.get_cmap('viridis')
fig = plt.figure(figsize=(8,6))
m1 = plt.scatter(X_train, y_train, color=cmap(0.9), s=10)
m2 = plt.scatter(X_test, y_test, color=cmap(0.5), s=10)
plt.plot(X, y_pred_line, color='black', linewidth=2, label='Prediction')
plt.show()
```
